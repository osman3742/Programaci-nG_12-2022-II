```python
import json
import requests
from pprint import pprint

response=requests.get("https://fp.josedomingo.org/lmgs/u05/ej2.json")
pendientes=json.loads(response.text)
pprint(pendientes)


def PruebasTotales(pendientes):
    titulos = []
    for prueba in pendientes:
      titulos.append(prueba["Titulo"])
    return titulos



def PruebasMasDeDosHoras(pendientes):
    titulos = []
    for prueba in pendientes:
        if prueba["Horas"]>2:
            titulos.append(prueba["Titulo"])
    return titulos

def UrlPruebasNoPresencial(pendientes):
    urls = []
    for prueba in pendientes:
        if prueba["TipoFormacion"]=="NoPresencial":
            urls.append(prueba["URL"])
    return urls

def PruebaTitulosProfesores(pendientes,id):
    datos = []
    for prueba in pendientes:
        if prueba["$id"]==id:
            datos.append(prueba["Titulo"])
            for profesor in prueba["Profesorado"]:
                datos.append(profesor["NombreCompleto"])
            return datos
    return datos

print("Ejercicio 1")
for titulo in PruebasTotales(pendientes):
  print(titulo)

print("Ejercicio 2")
for titulo in PruebasMasDeDosHoras(pendientes):
    print(titulo)

print("Ejercicio 3")
for url in UrlPruebasNoPresencial(pendientes):
    print(url)

print("Ejercicio 4")
id = input("Dame el id de una prueba:")
datos=PruebaTitulosProfesores(pendientes,id)
if len(pendientes)>0:
    print("Ti­tulo:",datos[0])
    print("Profesores")
    for profesor in datos[1:]:
        print(profesor)
else:
    print("No existe esa prueba de nivel")
```

```python
def InfoPrueba(pendientes):
    titulos = []
    profesores=[]
    for prueba in pendientes:
        titulos.append(prueba["Titulo"])
        lista_prof = []
        for profesor in prueba["Profesorado"]:
                lista_prof.append(profesor["NombreCompleto"])
        profesores.append(lista_prof)
    return zip(titulos,profesores)


print("Ejercicio 5")

for titulo,profesores in InfoPrueba(pendientes):
    print(titulo)
    print("Profesores:")
    for profesor in profesores:
        print(profesor)
      
```
